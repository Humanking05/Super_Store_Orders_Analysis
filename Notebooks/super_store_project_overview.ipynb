## Super Store Sales & Profit Analysis


## Project Objective

- Analyze Super Store sales and profits.

- Calculate key metric and non-metric KPIs.

- Include monthly and yearly sales trends for business planning.

- Provide insights to guide business decisions.

- Prepare the data for visualizations and dashboards.

## Dataset Overview

- Source: super_store_orders CSV file uploaded to Databricks.

- Key columns include:

- Order ID, Order Date, Ship Date

- Customer ID, Segment, Region

- Category, Sub-Category, Product Name

- Sales, Quantity, Profit

## Data Transformation Summary

- Converted text dates to real date format (Order Date, Ship Date).
 
- Calculated metric KPIs (sales, profit, AOV, profit margin).

- Calculated non-metric KPIs (best/worst categories, top region, top customer segment).
 
- Calculated monthly sales trend:

- Monthly: SUM(Sales) grouped by month

- Transformation was done entirely in SQL queries inside the super_store_orders_notebook.

- No additional tables were created; queries are run in place for KPI calculations.

## Metric KPIs

| KPI                       | Value                                 | Business Purpose                        |
| ------------------------- | ------------------------------------- | --------------------------------------- |
| Total Sales               | `SUM(Sales)`                          | Measure total revenue generated         |
| Total Profit              | `SUM(Profit)`                         | Assess bottom-line performance          |
| Average Order Value (AOV) | `SUM(Sales)/COUNT(DISTINCT Order ID)` | Understand customer purchasing behavior |
| Profit Margin (%)         | `(SUM(Profit)/SUM(Sales))*100`        | Evaluate efficiency and profitability   |


## Non-Metric KPIs

| KPI                           | Value                             | Business Purpose                             |
| ----------------------------- | --------------------------------- | -------------------------------------------- |
| Best-Performing Category      | `Category with highest sales`     | Identify top revenue-generating product line |
| Worst-Performing Sub-Category | `Sub-Category with lowest profit` | Highlight areas to optimize or remove        |
| Region With Most Orders       | `Region with highest order count` | Guide marketing and logistics focus          |
| Top Customer Segment          | `Segment generating most sales`   | Target promotions and loyalty programs       |


## Sales Trends
- Monthly Sales Trend

- Shows revenue patterns month by month.

- Helps plan promotions, inventory, and staffing.

## Scheduling
Scheduled the data with databricks to update everyday


## Insights & Recommendations

- Focus on high-performing categories for inventory planning and marketing.

- Investigate low-profit sub-categories to optimize pricing or remove unprofitable products.

- Increase marketing and logistics support in the top-performing region.

- Design loyalty programs or targeted promotions for the top customer segment.

- Use monthly sales trends to plan seasonal promotions, forecast demand, and allocate resources.
