```sql
-- Select all columns from the super_store_orders table
SELECT * 
FROM super_store_orders.default.super_store_orders
-- Limit the result to 10 rows
LIMIT 10;
```

```sql
SELECT
  *,
  TRY_TO_DATE(order_date, 'M/d/yyyy') AS order_date_converted, -- Convert order_date to date format
  TRY_TO_DATE(ship_date, 'M/d/yyyy') AS ship_date_converted    -- Convert ship_date to date format
FROM super_store_orders.default.super_store_orders
LIMIT 10;
```

```sql
-- Count total rows and missing values in order_date and ship_date
SELECT
  COUNT(*) AS total_rows, -- Total number of rows in the table
  count_if(order_date IS NULL) AS missing_order_date, -- Count of rows with missing order_date
  count_if(ship_date IS NULL) AS missing_ship_date -- Count of rows with missing ship_date
FROM super_store_orders.default.super_store_orders;
```

```sql
-- Select all columns from the super_store_orders table, removing duplicate rows
SELECT DISTINCT *
FROM super_store_orders.default.super_store_orders
LIMIT 10;
```

```sql
-- Calculate total sales from the super_store_orders table
SELECT
  SUM(sales) AS total_sales
FROM super_store_orders.default.super_store_orders;
```

```sql
-- Calculate total profit from the super_store_orders table
SELECT
  SUM(profit) AS total_profit
FROM super_store_orders.default.super_store_orders;
```

```sql
-- Calculate average order value from the super_store_orders table
SELECT
  AVG(sales) AS average_order_value
FROM super_store_orders.default.super_store_orders;
```

```sql
-- Calculate profit margin from the super_store_orders table
SELECT
  SUM(profit) / NULLIF(SUM(sales), 0) AS profit_margin
FROM super_store_orders.default.super_store_orders;
```

```sql
-- Calculate the best-performing category by total sales
SELECT
  category,
  SUM(sales) AS total_sales
FROM super_store_orders.default.super_store_orders
GROUP BY category
ORDER BY total_sales DESC;
```

```sql
-- Calculate worst-performing sub-category by total profit
SELECT
  sub_category,
  SUM(profit) AS total_profit
FROM super_store_orders.default.super_store_orders
GROUP BY sub_category
ORDER BY total_profit ASC
LIMIT 5;
```

```sql
-- Calculate the region with the highest order demand
SELECT
  region,
  COUNT(*) AS order_count
FROM super_store_orders.default.super_store_orders
GROUP BY region
ORDER BY order_count DESC;
```

```sql
-- Calculate the customer segment that generates the most sales
SELECT
  segment AS customer_segment,
  SUM(sales) AS total_sales
FROM super_store_orders.default.super_store_orders
GROUP BY segment
ORDER BY total_sales DESC;
```

```sql
-- Monthly sales trend
SELECT
  MONTH(TRY_TO_DATE(order_date, 'M/d/yyyy')) AS order_month,
  SUM(sales) AS monthly_sales
FROM super_store_orders.default.super_store_orders
GROUP BY
  MONTH(TRY_TO_DATE(order_date, 'M/d/yyyy'))
ORDER BY
  order_month;
```
